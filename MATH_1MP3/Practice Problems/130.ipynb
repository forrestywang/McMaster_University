(122) Make functions that takes in a square NumPy array as a parameter and performs the following actions using slicing:
1: Extract the i-th row of A
2: Extract the j-th column of B
3: Assume that the NumPy array is 3x3. Extract C00
4: Assume that the NumPy array is 3x3. Extract C01
5: Assume that the NumPy array is 3x3. Extract C02

```python
# Function(s):
def row_extractor(M, i):
    # Output:
    return M[i - 1]

def col_extractor(M, j):
    # Output:
    return M[:, j - 1]

def cofactor_1(M):
    # Output:
    return M[1:, 1:]

def cofactor_2(M):
    # Output:
    return M[1:, ::2]

def cofactor_3(M):
    # Output:
    return M[1:, :2]
```

(124) Make a function that takes in a 3x3 matrix as a NumPy array and returns its determinant using cofactor expansion.

```python
# Function(s):
def det_2x2(M):
    # Output:
    return M[0, 0] * M[1, 1] - M[0, 1] * M[1, 0]

def det_3x3(M):
    # Variable(s):
    det_array = np.array([det_2x2(cofactor_1(M)), - det_2x2(cofactor_2(M)), det_2x2(cofactor_3(M))])

    # Output:
    return np.dot(M[0], det_array)
```

(125) Make a function that takes in a 4x4 matrix as a NumPy array and returns its determinant using cofactor expansion.

```python
# Function(s):
def det_4x4(M):
    # Variable(s):
    cofact_1 = M[1:, 1:]
    cofact_2 = np.concatenate((M[1:, 0].reshape(3, 1), M[1:, 2:]), axis=1)
    cofact_3 = np.concatenate((M[1:, 0:2], M[1:, 3].reshape(3, 1)), axis=1)
    cofact_4 = M[1:, :3]

    # Calculations:
    det_array = np.array([det_3x3(cofact_1), - det_3x3(cofact_2), det_3x3(cofact_3), - det_3x3(cofact_4)])

    # Output:
    return np.dot(M[0], det_array)
```

(126) Make a function that returns a NumPy array of the values of g(x) = e^(0.2x) of numbers ∈ [0, 4] such that each number is 0.001 greater than the previous.

```python
# Function(s):
def g_array():
    # Output:
    return np.exp(- 0.2 * np.append(np.arange(0, 4, 0.001), 4))

# Output:
g_array()
```

(127) Make a function that returns a NumPy array of the values of f(x) = sin(2x) - 3 cos(x / 5) at 150 equally spaced values between 0 and π / 2.

```python
# Function(s):
def f(x):
    # Output:
    return np.sin(2 * x) - 3 * np.cos(x / 5)

def f_array():
    # Output:
    return f(np.linspace(0, np.pi / 2, 152))

# Output:
f_array()
```

(129) Make a function that returns the mean of the squares of the first 100 natural numbers.

```python
# Function(s):
def hundred_squares_mean():
    # Output:
    return np.mean(pow(np.arange(1, 101), 2))

# Output:
hundred_squares_mean()
```
