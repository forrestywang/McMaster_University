(41) Make a function that takes a non-empty array of numbers as a parameter and returns the smallest and largest value in the array.

```python
# Function(s):
def min_and_max(num):
    # Variable(s):
    num_min, num_max = num[0], num[0]

    # Logic
    for element in num:
        if element < num_min: num_min = element

        elif element > num_max: num_max = element

    # Output:
    return [num_min, num_max]
```

(42) Make a function that calculates ∑ m = 1 -> 600 of ∑ n = 10 -> 450 of m / n.

```python
# Function(s):
def double_sum():
    # Variable(s):
    sum = 0

    # Calculations:
    for m in range(1, 601):
        for n in range(10, 451): sum += m / n

    # Output:
    return sum

# Output:
print(double_sum())
```

(43) Make a function that prints "I love math” once, then “I hate math” once; then “I love math” two times, then “I hate math” once; then “I love math” three times, then “I hate math” once; then “I love math” four times, then “I hate math” once, and so on, until a total of 100 lines are printed.

```python
# Function(s):
def printing_strings():
    # Variable(s):
    count_1, count_2 = 0, 1

    # Output:
    while count_1 != 100:
        for i in range(count_2):
            print("I love math")
            count_1 += 1

            if count_1 == 100: break

        if count_1 == 100: break

        print("I hate math")
        count_1 += 1
        count_2 += 1

# Output:
printing_strings()
```

(44) Make a function that makes a list A ∋ ∀i ∈ [0, 10000], A[i] = sum of the digits in i.

```python
# Function(s):
def sum_digits():
    # Variable(s):
    A = []

    # Calculations:
    for i in range(10001):
        sum = 0

        for num in str(i): sum += int(num)

        A.append(sum)

    # Output:
    return A

print(sum_digits())
```

(45) Make a function that takes in a 3x3 matrix as a parameter and returns a new 3x3 matrix whose entry = (aij)³ - 7

```python
# Function(s):
def new_matrix(A):
    # Variable(s):
    B = []

    # Logic:
    for row in A:
        new_row = []

        for entry in row: new_row.append(entry ** 3 - 7)

        B.append(new_row)

    # Output:
    return B
```

(46) Make a function that takes in a 3x3 matrix as a parameter and returns the largest entry and the number of occurrences.

```python
# Function(s):
def maxEntry(A):
    # Variable(s):
    max = A[0][0]
    count = 1

    # Logic:
    for row in A:
        for entry in row:
            if entry > max:
                max = entry
                count = 1

            elif entry == max: count += 1

    # Output:
    return [max, count]
```

(47) Make a function that takes two 3x3 matrices A and B as parameters and returns A + B and 5A - 12B.

```python
# Function(s):
def matrix_addition(A, B):
    # Variable(s):
    sum_1, sum_2 = [], []

    # Logic:
    for row_num in range(3):
        row_1, row_2 = [], []

        for col_num in range(3):
            row_1.append(A[row_num][col_num] + B[row_num][col_num])
            row_2.append(5 * A[row_num][col_num] - 12 * B[row_num][col_num])

        sum_1.append(row_1)
        sum_2.append(row_2)

    # Output:
    return [sum_1, sum_2]
```

(48) Make a function that takes in a natural number as a parameter and determines if it is prime.

```python
# Function(s):
def is_prime(n):
    # Logic:
    if n == 1: return False

    for i in range(2, n):
        if n % i == 0: return False

    return True
```

(49) Make a function that returns a list of the first 100 prime numbers.

```python
# Function(s):
def hundred_primes():
    # Variable(s):
    primes_array = []
    n = 1

    # Logic:
    while len(primes_array) != 100:
        if is_prime(n): primes_array.append(n)

        n += 1

    # Output:
    return primes_array

# Output:
print(hundred_primes())
```

(50) Make a function that takes a natural number as a parameter and returns an array of all of its divisors including 1 and itself.

```python
# Function(s):
def all_divisors(n):
    # Variable(s):
    divisors_array = []

    # Logic:
    for i in range(1, n + 1):
        if n % i == 0: divisors_array.append(i)

    # Output:
    return divisors_array
```
