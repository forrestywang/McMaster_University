(51) Make a function that takes a speed in km/h as a parameter returns its speed in m/s and mph.

```python
# Function(s):
def speed_converter(x):
    # Output:
    print("m/s:", x * 5 / 18)
    print("mph:", x / 1.609)
```

(52) Make a function that takes in 3 positive real numbers as parameters and determines if they form a Pythagorean triplet.

```python
# Function(s):
def pythagorean_triplet(a, b, c):
    # Logic:
    if a ** 2 == b ** 2 + c ** 2 or b ** 2 == a ** 2 + c ** 2 or c ** 2 == a ** 2+ b ** 2: return True

    else: return False
```

(53) Make a function that takes in a 3x3 matrix as a parameter and returns its transpose.

```python
# Function(s):
def transpose(M):
    # Variable(s):
    A = []

    # Calculations:
    for col_num in range(3):
        transpose_row = []

        for row in M: transpose_row.append(row[col_num])

        A.append(transpose_row)

    # Output:
    return A
```

(54) Make a function that takes in a string as a parameter and prints the frequency of each vowel in it.

```python
# Function(s):
def num_vowels(string):
    # Variable(s):
    count_a, count_e, count_i, count_o, count_u = 0, 0, 0, 0, 0

    # Calculations:
    for letter in string:
        if letter.lower() == "a": count_a += 1
        if letter.lower() == "e": count_e += 1
        if letter.lower() == "i": count_i += 1
        if letter.lower() == "o": count_o += 1
        if letter.lower() == "u": count_u += 1

    print("a:", count_a)
    print("e:", count_e)
    print("i:", count_i)
    print("o:", count_o)
    print("u:", count_u)
```

(55) Make a function that computes the sum of the digits in each number in [1, 2,000,000].

```python
# Function(s):
def sum_digits_1():
    # Variable(s):
    total_digits = 0

    # Calculations:
    for num in range(1, 2000001): total_digits += len(str(num))

    # Output:
    return total_digits

# Output:
print(sum_digits_1())
```

(56) Make a function that returns a list of the first 10,000 natural numbers that are not divisible by 3 and do not contain the digit 3 and returns the 10,000 number in the list.

```python
# Function(s):
def subset():
    # Variable(s):
    subset = []

    # Calculations:
    for num in range(1, 10001):
        if num % 3 != 0 and "3" not in str(num): subset.append(num)

    # Output:
    return subset

# Output:
print(subset())
print()
print("Last number:", subset()[-1])
```

(57) Make a function that returns the total number of digits in the first 10,000 natural numbers and the number of times 7 appears.

```python
# Function(s):
def sum_digits_2():
    # Variable(s):
    total_digits = count_7 = 0

    # Calculations:
    for num in range(1, 10001):
        total_digits += len(str(num))

        for digit in str(num):
            if digit == "7": count_7 += 1

    # Output:
    print("Total digits:", total_digits)
    print("Total 7s:", count_7)

# Output:
sum_digits_2()
```

```python
# Function(s):
def sum_digits_3():
    # Variable(s):
    total_digits = 0

    # Calculations:
    for num in range(1, 10000):
        if "3" not in str(num): total_digits += len(str(num))

    # Output:
    print("Total digits:", total_digits)
    print("Total digits divisible by 3:", total_digits % 3 == 0)

# Output:
sum_digits_3()
```

(59) Make a function that computes the sum ∑ n = 2 -> 1000 (Fn / (Fn+1 * Fn−1)), where Fn is the nth Fibonacci number.

```python
# Function(s):
def fibonacci_sum(sum, f1, f2, count):
    # Calculations:
    while count <= 9999:
        count += 1
        f2 += f1
        f1 = f2 - f1

        sum += f1 / (f2 * (f2 - f1))

    # Output:
    return sum

# Variable(s):
sum = 0
f1, f2 = 1, 1
count = 1

# Output:
print(fibonacci_sum(sum, f1, f2, count))
```

(60) Make a function that returns an array of the first 10,000 Fibonacci numbers, an array of the last digit of each number, and an array of the number of digits in each number.

```python
# Function(s):
def fibonacci_arrays(f1, f2, fib_nums, last_digit, num_digit, count):
    # Calculations:
    while count <= 9998:
        count += 1
        f2 += f1
        f1 = f2 - f1

        fib_nums.append(f2)
        last_digit.append(f2 % 10)
        num_digit.append(len(str(f2)))

    # Output:
    return [fib_nums, last_digit, num_digit]

# Variable(s):
f1, f2 = 1, 1
fib_nums, last_digit, num_digit = [1, 1], [1, 1], [1, 1]
count = 1

# Output:
fibonacci_arrays(f1, f2, fib_nums, last_digit, num_digit, count)
print("First 10,000 Fibonacci numbers:", fib_nums)
print("Last digit of each number:", last_digit)
print("Number of digits in each number:", num_digit)
```
