(1) Make a function named spell_number that takes in a natural number ∈ [1, 9999] as a parameter and returns the number in English.

```python
# Function(s):
def spell_number(n):
    # Variable(s):
    name = ""
    units_digit = n % 10
    tens_digit = n % 100 // 10 * 10
    hundreds_digit = n % 1000 // 100 * 100
    thousands_digit = n // 1000 * 1000
    irregular_digit = n % 100

    # Dictionaries:
    UNITS_DIGIT_NAME = {1: "one", 2: "two", 3: "three", 4: "four", 5: "five", 6: "six", 7: "seven", 8: "eight", 9: "nine"}
    TENS_DIGITS_NAME = {10: "ten", 20: "twenty", 30: "thirty", 40: "forty", 50: "fifty", 60: "sixty", 70: "seventy", 80: "eighty", 90: "ninety"}
    HUNDREDS_DIGITS_NAME = {100: "one hundred", 200: "two hundred", 300: "three hundred", 400: "four hundred", 500: "five hundred", 600: "six hundred", 700: "seven hundred", 800: "eight hundred", 900: "nine hundred"}
    THOUSANDS_DIGITS_NAME = {1000: "one thousand", 2000: "two thousand", 3000: "three thousand", 4000: "four thousand", 5000: "five thousand", 6000: "six thousand", 7000: "seven thousand", 8000: "eight thousand", 9000: "nine thousand"}
    IRREGULAR_DIGITS_NAME = {11: "eleven", 12: "twelve", 13: "thirteen", 14: "fourteen", 15: "fifteen", 16: "sixteen", 17: "seventeen", 18: "eighteen", 19: "nineteen"}

    # Logic:
    if thousands_digit in THOUSANDS_DIGITS_NAME:
        name += THOUSANDS_DIGITS_NAME[thousands_digit] + " "

    if hundreds_digit in HUNDREDS_DIGITS_NAME:
        name += HUNDREDS_DIGITS_NAME[hundreds_digit] + " "

    if irregular_digit in IRREGULAR_DIGITS_NAME:
        name += IRREGULAR_DIGITS_NAME[irregular_digit]

    else:
        if tens_digit in TENS_DIGITS_NAME:
            name += TENS_DIGITS_NAME[tens_digit] + " "

        if units_digit in UNITS_DIGIT_NAME:
            name += UNITS_DIGIT_NAME[units_digit]

    # Output:
    return name

# Test case(s):
print(spell_number(9307))  # 'nine thousand three hundred seven'
print(spell_number(5004))  # 'five thousand four'
print(spell_number(1616))  # 'one thousand six hundred sixteen'
print(spell_number(13))  # 'thirteen'
print(spell_number(906))  # 'nine hundred six'
print(spell_number(67))  # 'sixty seven'
print(spell_number(632))  # 'six hundred thirty two'
print(spell_number(2))  # 'two'
print(spell_number(111))  # 'one hundred eleven'
```

(2) Make a function named count_two which takes in a string of the first 1 million decimals of pi, and returns the list how_many_times, which contains the number of times a two-digit number appears in the string. Print the two-digit number that appears the least and the most.

```python
# Import(s):
import requests
data_string = requests.get("https://ms.mcmaster.ca/lovric/1MP3/files/pi1million.txt").text

# Function(s):
def count_two():
    # Variable(s):
    how_many_times = [0] * 100

    # Logic:
    for start in range(len(data_string) - 1):
        end = start + 2

        how_many_times[int(data_string[start:end])] += 1

    # Output:
    return how_many_times

# Variable(s):
distribution_array = count_two()
max_occurrences = max(distribution_array)
min_occurrences = min(distribution_array)
max_value = distribution_array.index(max_occurrences)
min_value = distribution_array.index(min_occurrences)

# Output:
print("Most common two-digit number:", max_value, "with", max_occurrences, "occurrences")
print("Least common two-digit number:", min_value, "with", min_occurrences, "occurrences")
print("Two-digit distribution array:", distribution_array)
```

(3) Make a function that returns an array of all the unique 15-letter words in lines. Make another function that returns an array of all the unique words in lines. Print the length of both arrays.

```python
# Import(s):
import re
import urllib.request as ur
web_page = ur.urlopen('https://ms.mcmaster.ca/lovric/1MP3/files/TheGreatGatsby.txt')
lines = web_page.readlines()  # lines is a list of lines

# Cleaning the data:
for i in range(len(lines)):
    lines[i] = lines[i].decode("utf-8")  # lines is a list of lines stripped of all markup, new lines, etc
    lines[i] = re.sub("[^\w]", " ",  lines[i]).split()

# Function(s):
def unique_words_15():
    # Variable(s):
    unique_word = []

    # Logic:
    for sentence in lines:
        for word in sentence:
            if len(word) == 15 and word not in unique_word:
                unique_word.append(word)

    # Output:
    return unique_word

def unique_words():
    # Variable(s):
    unique_word = []

    # Logic:
    for sentence in lines:
        for word in sentence:
            if word not in unique_word:
                unique_word.append(word)

    # Output:
    return unique_word

# Variable(s):
all_15_letter_unique = unique_words_15()
all_unique = unique_words()

# Output:
print("The list of unique words is not accurate because contractions are counted as two \"words\". The part before the apostrophe is in the array, and the part after is also in the array. Hence, the number of unique words computed using this code will always be greater than or equal to the number of actual unique words.")
print("------------------------------------------------------------------------------------------------------------------------------------------")
print("Number of unique 15-letter words:", len(all_15_letter_unique))
print("Number of unique words:", len(all_unique))
print("All unique 15-letter words:", all_15_letter_unique)
print("All unique words:", all_unique)
```

(4) Make a function that takes in two dictionaries with numerical values as parameters, and returns a new dictionary that is the concatenation of the two, and if multiple values are mapped from the same key, then the values are summed.

```python
# Function(s):
def concatenate_dict(d1,d2):
    # Logic:
    for key, value in d2.items():
        if key not in d1:
            d1[key] = value

        else:
            d1[key] += value

    # Output:
    return d1

# Test case(s):
print(concatenate_dict({'a':1,'b':2},{'d':4}))  # {'a': 1, 'b': 2, 'd': 4}
print(concatenate_dict({'a':1,'b':2,'d':4,'m':4},{'d':2,'e':7,'m':11}))  # {'a': 1, 'b': 2, 'd': 6, 'm': 15, 'e': 7}
print(concatenate_dict({'a':12},{'a':2}))  # {'a': 14}
```

(5) Make a function that takes in two dictionaries as parameters, and returns a new dictionary according to the following rule:
a:b in d1 ∧ b:c in d2 ⇒ a:c in d3

```python
# Function(s):
def transitive_dict(d1,d2):
    # Variable(s):
    d3 = dict()

    # Logic:
    for key_1, value_1 in d1.items():
        if value_1 in d2:
            d3[key_1] = d2[value_1]

    # Output:
    return d3

# Test case(s):
print(transitive_dict({2:3, 8:19, 6:4, 5:3},{2:5, 4:3, 3:9}))  # {2: 9, 6: 3, 5: 9}
print(transitive_dict({2:3, 6:4, 5:3},{2:5, 6:9}))  # {}
print(transitive_dict({2:3, 3:3, 4:3},{3:2, 5:3, 6:9}))  # {2: 2, 3: 2, 4: 2}
```
